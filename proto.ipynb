```python
def process_listings(listings, window=7*24*60*60):
    lowest_price = None
    listings_updated = 0
    unique_retainers = {}
    t = int(time.time())
    for x in listings:
        if x["onMannequin"]:
            continue
        if t-x["lastReviewTime"] > window:
            continue
        if lowest_price is None or x["pricePerUnit"] < lowest_price:
            lowest_price = x["pricePerUnit"]
        listings_updated += 1
        unique_retainers[x["retainerID"]]=True

    return lowest_price, listings_updated, len(unique_retainers.keys())

process_listings(out["items"][str(out["itemIDs"][0])]["listings"])
```

```python
def process_sales(sales, window=7*24*60*60, max_count=None):
    #"Avg price per", "Listings sold", "Items Sold", "Unique Buyers",
    pass
    #filter mannequins, launderers
    t = time.time()
    listings_sold = 0
    items_sold = 0
    avg_price = [] #per listing, not per item in stack
    unique_buyers = {}
    for x in sales:
        if x["onMannequin"]:
            continue
        if int(t)-x["timestamp"] > window:
            # print(int(t)-x["timestamp"],window)
            continue
        unique_buyers[x['buyerName']] = True
        avg_price += x['pricePerUnit']
        items_sold += x["quantity"]
        listings_sold += 1

    avg = sum(avg_price)/len(avg_price) if len(avg_price)!= 0 else -1 

    return listings_sold, items_sold, avg, len(unique_buyers.keys())

process_sales(out["items"][str(out["itemIDs"][0])]["recentHistory"])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import argrelextrema

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 20
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Find local maxima in zs
local_maxima_indices = argrelextrema(zs, np.greater)[0]
print(local_maxima_indices)
# local_maxima_zs = zs[local_maxima_indices]

# Plot with a figure size of 20x20
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10, 10))
sc = ax.scatter(xs, ys, zs)

for x, y, z in zip(xs, ys, zs):
    label = ' (%d, %d, %d)' % (x, y, z)#, None)
    ax.text(x, y, z, label, None)
    ax.plot([x, x], [y, y], [z, 0], color='blue', label='Line to XY Plane')

# Add larger labels for axes
ax.set_xlabel('X Axis Label', fontsize=20)
ax.set_ylabel('Y Axis Label', fontsize=20)
ax.set_zlabel('Z Axis Label', fontsize=20)

ax.invert_zaxis()
plt.show()
```
